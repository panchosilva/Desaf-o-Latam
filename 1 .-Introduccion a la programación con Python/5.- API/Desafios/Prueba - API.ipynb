```python
#Inicio y fin de html

Inicio = "<html><head></head><body><ul>"
fin = "</ul></body></html>"

#Ejemplo de cómo debería quedar cada linea con cada link de cada imagen:
#<li><img src='http://mars.jpl.nasa.gov/msl-raw-images/msss/00025/mcam/0025MR0120007000E1_DXXX.jpg' width="100" height="100"><li>

def build_web_page(lista):
    string = "<html><head></head><body><ul>"
    for i in range(25):
        string += "<li><img src =\""+lista[i]+"\" width=\"100\" height=\"100\"><li>"
    string += "</ul></body></html>"
    return string
```
