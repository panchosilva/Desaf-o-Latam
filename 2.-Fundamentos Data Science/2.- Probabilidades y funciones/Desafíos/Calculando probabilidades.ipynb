# Desafío 4: Calcule la probabilidad de clasificación a la siguiente ronda

# a) Calcule la probabilidad de que un país pase a la siguiente ronda(clasifique) por continente.

Consejos:​ El problema puede ser resuelto de varias formas:

○ La manera más intuitiva corresponde a calcular la probabilidad en cada subconjunto creado en el problema 3.

○ Una forma alternativa es copiar todos los subsets creados en el problema 3 en una lista e iterar sobre dicha lista.

○ Una forma más elegante consiste en utilizar una operación de agrupación por continente, y luego calcular la probabilidad.

```python
# Voy a usar los objetos ya creados en Desafío 3

def prob_por_continent(dataframe):
  #Guardo los distintos continentes en un array
  continentes = dataframe['continent'].unique()

  #Itero sobre el array de continentes
  for i in continentes:
    df_0 = dataframe[dataframe['continent'] == i] #Filtro un df auxiliar para calcule la probabilidad de interés por cada continente
    probabilidad = np.sum(df_0['clasificado'])/len(df_0[['clasificado']].dropna()) # Generalizo quitando los "NaN" que pudiesen estar, para calcular bien la probabilidad
    print(f"La probabilidad de que un país del continente '\033[1m'{i}'\033[0m' clasifique es de '\033[1m'{probabilidad} '\033[0m'") #Imprime el resultado por cada continente
  
prob_por_continent(df)
```

# Desafío 5: Probabilidades de Victoria y Derrota

● Se pide utilizar​ value_counts ​para calcular la probabilidad de que un continente contenga países  con o sin victorias.

● Para ello cree una columna adicional llamada​ 'ha_ganado' que contenga un 0 si es que un país no tiene juegos ganados y un 1 si es que ha ganado al menos un juego.

● Sugerencias:​ Para crear dicha columna puede revisar: np.where​ de Numpy, o pd.where de Pandas. Además, puede utilizar los mismos procedimientos sugeridos en el Desafío 4.

```python
df['ha_ganado'] = np.where(df['juegos_ganados']>0,1,0)

new_df = pd.DataFrame(df.groupby(by = "continent")['ha_ganado'].value_counts('%')).rename(columns = {'ha_ganado':'porcentaje_interés'}).reset_index()
new_df
```

```python
new_df = new_df[new_df['ha_ganado'] == 1 ].drop(columns="ha_ganado")
new_df = new_df.sort_values(by = 'porcentaje_interés',ascending = False).reset_index(drop = True)
new_df
```

# En base al resultado del procedimiento recién descrito, responda lo siguiente:

```python
print(f"Según el resultado en a), el continente que tiene la mayor probabilidad de tener países con al menos 1 victoria es {new_df['continent'][0]} con probabilidad = a {new_df['porcentaje_interés'][0]}")
```
