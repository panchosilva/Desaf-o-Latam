# Desafío 5: Utilizando pandas y numpy

```python
import pandas as pd
import numpy as np

df = pd.read_csv("flights.csv")

print(df.head())
```

```python
print(df.tail())
```

```python
df.year.describe()
```

```python
df.year.value_counts()
```

```python
df.month.value_counts()
```

```python
variable_a = df.head(15)
variable_b = df.tail(15)
variable_c = variable_a.append(variable_b)
```

```python
variable_c
```

```python
#Datos de passengers
print(f"La media de la columna passengers es {np.mean(df.passengers)}")
print(f"La mediana de la columna passengers es {np.median(df.passengers)}")
print(f"La desviación estándar de la columna passengers es {np.std(df.passengers)}")
print()

#Datos de variable_c
print(f"La media de las últimas y primeras 15 observaciones es {np.mean(variable_c.passengers)}")
print(f"La mediana de las últimas y primeras 15 observaciones es {np.median(variable_c.passengers)}")
print(f"La desviación estándar de las últimas y primeras 15 observaciones es {np.std(variable_c.passengers)}")

print("""
Comentario: La cantidad de pasajeros ha aumetado durante el pasar del tiempo, y si se toman los extremos (primeros y últimos 15 meses),
es de esperar que la media aumente, la mediana curiosamente disminuya y que la desviación estándar sea más grande que tomando passengers completo
""")

import matplotlib.pyplot as plt
df.groupby(by = "year")['passengers'].plot()
plt.xlabel("Número de mes desde el inicio hasta el mes 143")
plt.ylabel("Cantidad de pasajeros")
plt.title("Evolución de los últimos 143 meses de datos (color por año)")
plt.show()

print("También se nota claramente una ciclicidad en la cantidad de pasajeros, va pasando por temporadas altas y bajas :) ")
```
