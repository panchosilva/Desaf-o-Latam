# Ejercicio 1: Generación Artificial de Datos

```python
import random
```

```python
def create_random_row(i):
    # simulamos la columna edad
    age = random.randint(18, 90)
    # simulamos la columna ingreso
    income = random.randrange(10000, 1000000, step=1000)
    # simulamos la situación laboral
    employment_status = random.choice(['Unemployed', 'Employed'])
    # simulamos si es que tiene deuda o no
    debt_status = random.choice(['Debt', 'No Debt'])
    # simulamos si es que se cambió recientemente o no
    churn_status = random.choice(['Churn', 'No Churn'])
    return age, income, employment_status, debt_status, churn_status
```

## a) Replique la función 10 millones de veces y preservarla en un objeto

```python
objeto = list(map(create_random_row, range(10000000)))

objeto[0:10]
```

```python
random_database = objeto
```

## e) Estime la media, la varianza, el mínimo y el máximo de la edad para ambos subsets, sin utilizar librerías externas.

```python
def media_edad(obj):
    a = list(map(lambda x: x[0], obj))
    return sum(a)/len(a)

def varianza_edad(obj):
    a = list(map(lambda x: x[0], obj))
    media = media_edad(obj)
    return sum(map(lambda x: pow(x-media,2), a))/len(a)

def minimo_edad(obj):
    return min(list(map(lambda x: x[0], obj)))

def maximo_edad(obj):
    return max(list(map(lambda x: x[0], obj)))

print(f"""
Datos de Churn relacionado a la edad:
La media: {round(media_edad(churns),2)}
varianza: {round(varianza_edad(churns),2)}
mínimo: {round(minimo_edad(churns),2)}
máximo: {round(maximo_edad(churns),2)}""")

print(f"""
Datos de NO_Churn relacionado a la edad:
La media: {round(media_edad(no_churns),2)}
varianza: {round(varianza_edad(no_churns),2)}
mínimo: {round(minimo_edad(no_churns),2)}
máximo: {round(maximo_edad(no_churns),2)}""")
```
